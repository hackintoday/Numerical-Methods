```python
import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(threshold=np.inf)

a= np.zeros((100,100), dtype=int)

for r in (-99, 0):
    for i in (-99, 0):
        if((1+r + r**2 - i**2)**2 + (i + 2*r*i)**2 < 1):
            a[r][i] = 1
        else:
            a[r][i] = 0
            
```
